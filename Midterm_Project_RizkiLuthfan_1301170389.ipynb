```python
data2 = r'Hang_Seng.csv'
data2 = pd.read_csv(data2)
data2.head()
```

```python
data2['Date'] = pd.to_datetime(data2['Date'])
data2.head(2)
```

```python
data2["Date"].max() - data2["Date"].min()
```

```python
data2.iloc[-90:].describe().astype(int)
```

```python
data2.index = data2["Date"]
```

```python
data2["Adj Close"].plot(figsize = (15, 8))
plt.show()
```

```python
data2["DailyPerchantageChange"] = data2["Adj Close"].pct_change()*100
data2.head()
```

```python
data2.dropna(axis = 0, inplace = True)
```

```python
data2["DailyPerchantageChange"].plot(figsize = (12, 6), fontsize = 12)
```

```python
data2["DailyPerchantageChange"].hist(bins = 50, figsize = (10,5))
plt.xlabel("Daily")
plt.ylabel("Frekuensi")
plt.show()

data2.DailyPerchantageChange.describe() #Data statistik
```

```python
def trends(x):
  if x > -0.5 and x <= 0.5:
    return "Slight or No change"
  elif x > 0.5 and x <= 1:
    return "Slight Positive"
  elif x > -1 and x <= -0.5:
    return "Slight Negative"
  elif x > 1 and x <= 3:
    return "Positive"
  elif x > -3 and x <= 11:
    return "Negative"
  elif x > 3 and x <= 7:
    return "Among top gainers"
  elif x > -7 and x <= -3:
    return "Among top losers"
  elif x > 7:
    return "Bull run"
  elif x <= -7:
    return "Bear drop"

data2["trends"] = data2["DailyPerchantageChange"].apply(lambda x:trends(x))
data2.head()
```

```python
pie_data = data2.groupby('trends')
pie_label = sorted([data2.loc[:, 'trends'].unique()])
plt.pie(pie_data['trends'].count(), autopct = '%1.1f%%', radius = 2)


plt.show()
```

```python
hasil = pie_data['trends'].count()
hasil
```
